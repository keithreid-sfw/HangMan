```python
def build_partial_word(word, right_so_far):
    partial_word = word
    return partial_word



def test_build_partial_word():
    word     = "ehfehveh" # string
    
    right_so_far = ['e']
    partial_word = build_partial_word(right_so_far, word)
    assert(partial_word == e__e__e_), "handling e wrong"
    print("build partial word handles e right")

test_build_partial_word()
    
```

```python
# s.3 view


def greeting():
    print("\nWelcome to Hangman\n")
    
def print_gibbet(wrong_so_far):
    gibbets = get_gibbets()
    print(gibbets[wrong_so_far])

#s.4 control

def invite_letter():
    new_letter = input("Choose a letter please:")
    return new_letter

#s.5 blocks

def intro():
    greeting()
    gibbets     = get_gibbets()
    chances     = len(gibbets)
    print("There are", chances, "stages in the Hang Man drawing.")
                
    words       = get_words()
    word        = random.choice(words)
    letters     = len(word)
    print("The Word has", letters, "letters.")
    
    dummy_word  = "_"*letters
    print("Dummy word:\t", dummy_word)

def game_loop(word):
    alphabet     = get_alphabet()
    gibbets      = get_gibbets()
    chances      = len(gibbets)
    
    wrong_so_far = 0
    right_so_far = []
    
    while wrong_so_far < chances-1:

        letter = invite_letter()
        print("You chose", letter)
        if letter in word:
            print("correct")
            right_so_far.append(letter)
            print("""
            Right so far:""", right_so_far)
            partial_word = build_partial_word(right_so_far, word)
            
        else:
            print("incorrect")
            wrong_so_far = wrong_so_far + 1
            print_gibbet(wrong_so_far)
```
